```python
class Team:
    def __init__(self, name):
        self.name = name
        self.players = {}

    def addPlayer(self, *players):
        for player in players:
            if player.role != None:
                if self.players.get(player.role) != None:
                    self.players[player.role].append([player.name, player.number])
                else:
                    self.players[player.role] = []
                    self.players[player.role].append([player.name, player.number])
                print(f"{player.name} added in team")
            else:
                print('A player cannot be added without a role.')

    def showPlayers(self):
        for key in self.players.keys():
            print(f"{key}:")
            for id in range(len(self.players[key])):
                player = self.players[key][id]
                print(f"Player name: {player[0]} Jersey No: {player[1]}")
        



class Player:
    def __init__(self, *details):
        if len(details) == 3:
            self.name = details[0]
            self.role = details[1]
            self.number = details[2]
        elif len(details) == 2:
            self.name = details[0]
            self.number = details[1]
            self.role = None
    
    def setRole(self, role):
        self.role = role
        

team = Team("Bangladesh")
p1 = Player("Mahmudullah", "Batsman",30)
print("=================================")
team.addPlayer(p1)
print("=================================")
team.showPlayers()
print("=================================")
p2 = Player("Mustafizur Rahman", "Bowler",90)
p3 = Player("Tamim Iqbal", 28)
print("=================================")
team.addPlayer(p2,p3)
print("=================================")
team.showPlayers()
print("=================================")
p3.setRole("Batsman")
print("=================================")
team.addPlayer(p3)
print("=================================")
team.showPlayers()
print("=================================")
p4 = Player("Mushfiqur Rahim", "Batsman",15)
p5 = Player("Taskin Ahmed", "Bowler",3)
print("=================================")
team.addPlayer(p4,p5)
print("=================================")
team.showPlayers()
```
