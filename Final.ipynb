## Question 1:
Write a python program that will take input from the user until the user 
gives “STOP” as input. In every line the user will provide input as follow:
[Brand Name] [PRICE]  
Your task is to create a dictionary where the brand’s name will be key and 
the value will be a list of that brand’s watches ID. ID generate process: 
FirstTwoLetterOfBrand_Price_inputNo. Finally print the dictionary.  

Sample Input 1:  
LENOVO 52000  
DELL 65000  
ASUS 80000  
LENOVO 90000  
DELL 70000   
STOP  
Sample Output 1:  
{'LENOVO': ['LE_52000_1', 'LE_90000_4'], 'DELL':   
['DE_65000_2', 'DE_70000_5'], 'ASUS': ['AS_80000_3']}  
  
Sample Input 2:  
MACBOOK 120000  
HP 60000  
DELL 55000  
HP 70000  
MACBOOK 150000  
STOP  
Sample Output 2:  
{'MACBOOK': ['MA_120000_1', 'MA_150000_5'], 'HP':   
['HP_60000_2', 'HP_70000_4'], 'DELL': ['DE_55000_3']}  


```python
result = {}
count = 1
while True:
    info = input("Enter information:").split(' ')
    if info[0] == 'STOP':
        break
    id = info[0][:2] + "_" + info[1] + "_" + str(count)
    if result.get(info[0]) != None:
        result[info[0]].append(id)
    else:
        result[info[0]] = []
        result[info[0]].append(id)
    count += 1

print(result)
```

```python
class Person:
  def __init__(self, name, contact):
    self.name = name
    self.contact = contact
 
  def __str__(self):
    s = f"Name: {self.name}\nContact: {self.contact}"
    return s


class Doctor(Person):
    def __init__(self, name, contact, speciality, hospital):
        super().__init__(name, contact)
        self.speciality= speciality
        self.hospital = hospital
        self.patient_list = {}
    
    def add_patients(self, *info):
        for patient in info:
            location = patient.split(' ')[0]
            if self.patient_list.get(location) != None:
                self.patient_list[location].append(patient)
            else:
                self.patient_list[location] = []
                self.patient_list[location].append(patient)
    
    def show_patient_list(self, location):
        if self.patient_list.get(location) != None:
            print(f'Patients from {location}:')
            for patiens in self.patient_list[location]:
                print(patiens)
        else:
            print(f'{self.name} has not patients from {location}')
        
    def __str__(self):
        output = "Doctor Details:\n"
        output += super().__str__() + '\n'
        output += 'Speciality: ' + self.speciality + '\n'
        output += "Hospital: " + self.hospital + '\n'
        output += "Patients List:" + '\n'
        output += str(self.patient_list)
        return output

class Engineer(Person):
    def __init__(self, name, contact, field):
        super().__init__(name, contact)
        self.field = field
        self.project_list = {}
    
    def add_projects(self, *info):
        for project in info:
            location = project.split(' ')[-1]
            if self.project_list.get(location) != None:
                self.project_list[location].append(project)
            else:
                self.project_list[location] = []
                self.project_list[location].append(project)

    def __str__(self):
        output = 'Engineer Details:\n'
        output += super().__str__() + '\n'
        output += f'Field: {self.field} \n'
        output += 'Project List: '
        output += str(self.project_list)
        return output

 
 
d = Doctor("Dr Strange", "4585632", "Neurosurgery", "DMC")
d.add_patients("Dhaka Abdur Rahim", "Dhaka John Cena", "Narayanganj John Wick", "Comilla CM Punk")
print("1.=========================================================")
print(d)
print("2.=========================================================")
d.show_patient_list("Dhaka")
print("3.=========================================================")
d.show_patient_list("Rajshahi")
print("4.=========================================================")
d.add_patients("Narayanganj Naruto Uzumaki")
print("5.=========================================================")
print(d)
print("6.=========================================================")
eng = Engineer("Howard Wolowitz", "982685", "Mechanical")
eng.add_projects("Mongol Tori Bangladesh", "Duburi Bangladesh", "Curiosity Rover USA", "Perseverence USA")
print("7.=========================================================")
print(eng)
```
